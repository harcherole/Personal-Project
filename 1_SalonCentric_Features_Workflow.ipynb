```python
date_parser = lambda c: pd.to_datetime(c, format='%Y/%m/%d %H:%M:%S', errors='coerce')
## This data contains only the header
header = pd.read_csv(path + '/' + file_with_header, sep=';')
## This data contains a csv file with 46 columns
csv_with_46_columns = pd.read_csv(path + '/' + file_with_46_columns, sep=';',
                            header=None,
                            parse_dates=[0, 4,24, 25],
                            date_parser = date_parser)
## This data contains a csv file with 48 columns
csv_with_48_columns = pd.read_table(path + '/' + file_with_48_columns, sep=';',
                                 header=None,
                                 parse_dates=[0, 4,24, 25],
                                 date_parser = date_parser,
                                 encoding='latin-1', 
                                 engine='python')
```

## This function will generate the dataframe with 46 columns

```python
def generateDataFrame(csv_with_46_columns, csv_with_48_columns):
    # The first CSV file contains 46 columns and the Second contains 48 columns
    # We drop two columns from the CSV file with 48 columns 
    # columns =[31,32] are empty because of initial data structure problem and will be dropped
    columns =[31,32]
    df_drop_2_columns = csv_with_48_columns.drop(columns, axis=1)
    # Adding header in this two dataframe
    df_drop_2_columns.columns = header.columns
    csv_with_46_columns.columns = header.columns
    # Concatenation of two dataframe
    frames = [df_drop_2_columns, csv_with_46_columns]
    df = pd.concat(frames)
    return df
```

```python
def CreateAllFeature_first_part(df):
    # We take a subset of our data frame, with only the needed variables
    # Each order :
    # - is associated with a unique Billing Document Number, 
    # - is placed by a Sold_to_code, 
    # - on a Shipment_Billing_Current_Analysis_Date
    # - has products which belongs to Material Groups & SubAxis
    # We sort the orders by increasing Shipment_Billing_Current_Analysis_Date and remove duplicates
    new_feature_1 = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 'Billing_Document_Number']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_1['previous_order_date'] = new_feature_1.groupby(['Sold_to_code'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # print(new_feature_1['previous_order_date'])
    # print(new_feature_1['Shipment_Billing_Current_Analysis_Date'])
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_1['lag'] = (new_feature_1['Shipment_Billing_Current_Analysis_Date'] - new_feature_1['previous_order_date']).dt.days
    # We groupb each 
    new_feature_1 = new_feature_1.groupby(["Sold_to_code"]).agg({'lag': ['min', 'max', 'mean', 'std']}).reset_index()
    new_feature_1.columns = [' '.join(col).strip() for col in new_feature_1.columns.values]
    # Rename feature 1
    new_feature_1.columns = ['Sold_to_code','min_time_diff','max_time_diff','avg_time_diff','std_time_diff']
    
    
    new_feature_2_MG = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 
                               'Billing_Document_Number', 'material_group_Description']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date', 
                     'material_group_Description']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_2_MG['previous_order_date'] = new_feature_2_MG.groupby(['Sold_to_code', 
                     'material_group_Description'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_2_MG['lag'] = (new_feature_2_MG['Shipment_Billing_Current_Analysis_Date'] - new_feature_2_MG['previous_order_date']).dt.days
    # We group each by customer and Material Group
    new_feature_2_MG = new_feature_2_MG.groupby(['Sold_to_code', 'material_group_Description']).agg({'lag': ['mean']}).reset_index()
    # We merge the two levels of column names
    new_feature_2_MG.columns = [' '.join(col).strip() for col in new_feature_2_MG.columns.values]
    # We pivot the table to get our MG in columns
    new_feature_2_MG = new_feature_2_MG.pivot_table(index=['Sold_to_code'],
                                                   columns=['material_group_Description'],
                                                   values='lag mean').reset_index()
    # We merge the two levels of column names
    new_feature_2_MG.columns = [''.join(col).strip() for col in new_feature_2_MG.columns.values]
    # Drop 2 columns
    new_feature_2_MG = new_feature_2_MG.drop(['Shopping Bags'], axis=1)

    # put columns names in lower function
    new_feature_2_MG.columns = ['Sold_to_code','accessory_promo_pos','allowance_funded_pac','bom_promo_no_fg','bom_sample_ppk_no_fg','bags',
                                'bandedpack_promo_gwp','box','displays_units_parts','fg_with_other_pack_i','financial_products','finished_goods',
                                'gwp','master_shipper_assor','other_pos','packaging_materials','partner_brand_fg','printed_materials','samples','t_shirts','testers','wip_basic_size']


    # Rename feature 2 Material Group without using first column
    new_names = [(i,i+'_time_reorder_avg') for i in new_feature_2_MG.iloc[:, 1:].columns.values]
    new_feature_2_MG.rename(columns = dict(new_names), inplace=True)
    
    
    new_feature_2_SA = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 
                               'Billing_Document_Number', 'Sub_Axe']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date', 
                     'Sub_Axe']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_2_SA['previous_order_date'] = new_feature_2_SA.groupby(['Sold_to_code', 
                     'Sub_Axe'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_2_SA['lag'] = (new_feature_2_SA['Shipment_Billing_Current_Analysis_Date'] - new_feature_2_SA['previous_order_date']).dt.days
    # We group each by customer and SubAxis
    new_feature_2_SA = new_feature_2_SA.groupby(['Sold_to_code', 'Sub_Axe']).agg({'lag': ['mean']}).reset_index()
    # We merge the two levels of column names
    new_feature_2_SA.columns = [' '.join(col).strip() for col in new_feature_2_SA.columns.values]
    # We pivot the table to get our MG in columns
    new_feature_2_SA = new_feature_2_SA.pivot_table(index=['Sold_to_code'],
                                                   columns=['Sub_Axe'],
                                                   values='lag mean').reset_index()
    # We merge the two levels of column names
    new_feature_2_SA.columns = [''.join(col).strip() for col in new_feature_2_SA.columns.values]

    # put columns names in lower function
    new_feature_2_SA.columns = ['Sold_to_code','haire_care','haire_color','others','styling']

    # Rename feature 2 SubAxis without using first column
    new_names_SubAxis = [(i,i+'_time_reorder_avg') for i in new_feature_2_SA.iloc[:, 1:].columns.values]
    new_feature_2_SA.rename(columns = dict(new_names_SubAxis), inplace=True)
    
    
    
    
    # merge all features
    data_1 = new_feature_1.merge(new_feature_2_MG,on='Sold_to_code').merge(new_feature_2_SA,on='Sold_to_code')
    return data_1
```

```python
def CreateAllFeature_second_part(df):
    # We take a subset of our data frame, with only the needed variables
    # Each order :
    # - is associated with a unique Billing Document Number, 
    # - is placed by a Sold_to_code, 
    # - on a Shipment_Billing_Current_Analysis_Date
    # - has products which belongs to Material Groups & SubAxis
    # We sort the orders by increasing Shipment_Billing_Current_Analysis_Date and remove duplicates
    new_feature_1 = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 'Billing_Document_Number']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_1['previous_order_date'] = new_feature_1.groupby(['Sold_to_code'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # print(new_feature_1['previous_order_date'])
    # print(new_feature_1['Shipment_Billing_Current_Analysis_Date'])
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_1['lag'] = (new_feature_1['Shipment_Billing_Current_Analysis_Date'] - new_feature_1['previous_order_date']).dt.days
    # We groupb each 
    new_feature_1 = new_feature_1.groupby(["Sold_to_code"]).agg({'lag': ['min', 'max', 'mean', 'std']}).reset_index()
    new_feature_1.columns = [' '.join(col).strip() for col in new_feature_1.columns.values]
    # Rename feature 1
    new_feature_1.columns = ['Sold_to_code','min_time_diff','max_time_diff','avg_time_diff','std_time_diff']
    
    # feature 5
    df['qty*val*div'] = df.Qty_BL_CM_Gross*df.Amt_BL_CM_Gross*df.divertible
    df['qty*val'] = df.Qty_BL_CM_Gross*df.Amt_BL_CM_Gross
    df['qty*div'] = df.Qty_BL_CM_Gross*df.divertible
    df[['Qty_BL_CM_Gross','Amt_BL_CM_Gross','divertible','qty*val*div','qty*val', 'qty*div']].head()
    grouped5 = df.groupby('Sold_to_code')
    interdf51 = grouped5['qty*val*div','qty*val'].sum()
    interdf51['avg_div_in_value'] = interdf51['qty*val*div']/interdf51['qty*val']
    interdf51 = interdf51.reset_index()
    feature5_1 = interdf51[['Sold_to_code','avg_div_in_value']]
    interdf52 = grouped5['qty*div','Qty_BL_CM_Gross'].sum()
    interdf52['avg_div_in_qty'] = interdf52['qty*div']/interdf52['Qty_BL_CM_Gross']
    interdf52 = interdf52.reset_index()
    feature5_2 = interdf52[['Sold_to_code','avg_div_in_qty']]
    df5_3 = df[['Sold_to_code','Billing_Document_Number','Amt_BL_CM_Gross','divertible']]
    df5_3['div*Amt'] = df5_3.loc[:,'divertible']*df5_3.loc[:,'Amt_BL_CM_Gross']
    grouped5_3 = df5_3.groupby(['Sold_to_code','Billing_Document_Number']).sum()
    grouped5_3['max_prop_val_div'] = grouped5_3['div*Amt']/grouped5_3['Amt_BL_CM_Gross']
    inter5_3 = grouped5_3.reset_index()
    feature5_3 = inter5_3.groupby(['Sold_to_code']).max()
    feature5_3 = feature5_3['max_prop_val_div'].reset_index()
    df5_4 = df[['Sold_to_code','Billing_Document_Number','Qty_BL_CM_Gross','divertible']]
    df5_4['div*Qty'] = df5_4.loc[:,'divertible']*df5_4.loc[:,'Qty_BL_CM_Gross']
    grouped5_4 = df5_4.groupby(['Sold_to_code','Billing_Document_Number']).sum()
    grouped5_4['max_prop_qty_div'] = grouped5_4['div*Qty']/grouped5_4['Qty_BL_CM_Gross']
    inter5_4 = grouped5_4.reset_index()
    feature5_4 = inter5_4.groupby(['Sold_to_code']).max()
    feature5_4 = feature5_4['max_prop_qty_div'].reset_index()
    feature55 = pd.merge(feature5_1,feature5_2)
    feature56 = pd.merge(feature5_3,feature5_4)
    feature5 = pd.merge(feature55, feature56)
    # Rename feature 5
    feature5.columns = ['Sold_to_code','amt_divertible_in_avg','qty_divertible_in_avg','amt_divertible_prop_max','qty_divertible_prop_max']
    
    # feature 6
    df['diff_date_first_Now_td'] = (df.Shipment_Billing_Current_Analysis_Date - df.SC_First_Ship_Date)
    df['diff_date_first_Now'] = df['diff_date_first_Now_td'].dt.days
    #df['diff_date_first_Now'].head()
    grouped6 = df.groupby('Sold_to_code')
    feature_6 = grouped6['diff_date_first_Now'].agg(['min','max', 'mean']).reset_index()
    # Rename feature 6
    feature_6.columns = ['Sold_to_code','age_prod_avg','age_prod_max','age_prod_min']
    
    # feature 7
    df7_1 = df[['SC_First_Ship_Date','Shipment_Billing_Current_Analysis_Date','Sold_to_code','Amt_BL_CM_Gross']]
    grouped7_1 = df7_1.groupby(['Sold_to_code']).max()
    grouped7_1['age_most_prod_val'] = (grouped7_1.Shipment_Billing_Current_Analysis_Date - grouped7_1.SC_First_Ship_Date).dt.days
    grouped7_1 = grouped7_1.reset_index()
    df7_2 = df[['SC_First_Ship_Date','Shipment_Billing_Current_Analysis_Date','Sold_to_code','Qty_BL_CM_Gross']]
    grouped7_2 = df7_2.groupby(['Sold_to_code']).max()
    grouped7_2['age_most_prod_qty'] = (grouped7_2.Shipment_Billing_Current_Analysis_Date - grouped7_2.SC_First_Ship_Date).dt.days
    grouped7_2 = grouped7_2.reset_index()
    feature7 = pd.merge(grouped7_1, grouped7_2)
    feature7 = feature7[['Sold_to_code','age_most_prod_val','age_most_prod_qty']]
    # Rename feature 7
    feature7.columns = [['Sold_to_code','amt_age_most_prod_avg','qty_age_most_prod_avg']]    
    
    # feature 8
    ## Nb distinct 'ship' adresses for not STORE
    split8 = df[df.distribution_channel_description != 'STORE'].groupby('Sold_to_code')
    disct_ship_not_store = split8.street_address.nunique()
    feature8 = pd.DataFrame(disct_ship_not_store).reset_index()
    all_stc = set(new_feature_1.Sold_to_code) # all data with column name Sold_to_code
    some_stc_ft_8 = set(disct_ship_not_store.index) # data with 1
    missing_ft_8 = all_stc - some_stc_ft_8 # missing data
    # je remplace des valeurs manquantes par des zeros Feature 8
    S_ft_8 = list(missing_ft_8)
    L_ft_8 = []
    for i in range(len(S_ft_8)):
        if S_ft_8[i] > 1:
            L_ft_8.append(0)
    #L_ft_8
    # je rajoute des zero remplacs prcdemment dans some_stc_ft_8
    List_feature_8_column_address = list(feature8['street_address'])
    List_feature_8_column_address.extend(L_ft_8)
    # je rajoute les Sold_to_code qui manquent dans features 8 colonne Sold_to_code
    List_feature_8_column_Sold = list(feature8['Sold_to_code'])
    List_feature_8_column_Sold.extend(S_ft_8)
    # DataFrame
    Feature_8 = pd.DataFrame({'Sold_to_code': List_feature_8_column_Sold, 
                          'street_address_channel_SC': List_feature_8_column_address})
    
    
    # feature 9
    ## Nb distinct 'ship' adresses for STORE
    split9 = df[df.distribution_channel_description == 'STORE'].groupby('Sold_to_code')
    disct_ship_store = split9.street_address.nunique()
    feature9 = pd.DataFrame(disct_ship_store).reset_index()
    
    all_stc = set(new_feature_1.Sold_to_code) # all data with column name Sold_to_code
    some_stc_ft_9 = set(disct_ship_store.index) # data with 1
    missing_ft_9 = all_stc - some_stc_ft_9 # missing data
    # je remplace des valeurs manquantes par des zeros Feature 9
    S_ft_9 = list(missing_ft_9)
    L_ft_9 = []
    for i in range(len(S_ft_9)):
        if S_ft_9[i] > 1:
            L_ft_9.append(0)
    #L_ft_9
    # je rajoute des zero remplacs prcdemment dans some_stc_ft_9
    List_feature_9_column_address = list(feature9['street_address'])
    List_feature_9_column_address.extend(L_ft_9)
    #List_feature_9_column_address
    # je rajoute les Sold_to_code qui manquent dans features 9 colonne Sold_to_code
    List_feature_9_column_Sold = list(feature9['Sold_to_code'])
    List_feature_9_column_Sold.extend(S_ft_9)
    #List_feature_9_column_Sold
    # DataFrame
    Feature_9 = pd.DataFrame({'Sold_to_code': List_feature_9_column_Sold, 
                          'street_address_channel_not_SC': List_feature_9_column_address})
    
    # feature 10
    df10_1 = df[['Sold_to_code','distribution_channel_description','Amt_BL_CM_Gross']]
    grouped10_1 = df10_1.groupby(['Sold_to_code','distribution_channel_description']).sum()
    grouped10_1 = grouped10_1.reset_index()
    grouped10_1_pivot = grouped10_1.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Amt_BL_CM_Gross').fillna(0)
    grouped10_1_pivot.columns = [''.join(col).strip() for col in grouped10_1_pivot.columns.values]
    grouped10_1_pivot = grouped10_1_pivot.reset_index()
    grouped10_1_pivot['prop_val_B2B'] = grouped10_1_pivot.B2B/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot['prop_val_STORE'] = grouped10_1_pivot.STORE/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot['prop_val_STREET'] = grouped10_1_pivot.STREET/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot = grouped10_1_pivot.reset_index()
    grouped10_1_pivot_prop_vals = grouped10_1_pivot[['Sold_to_code','prop_val_B2B','prop_val_STORE','prop_val_STREET']]
    df10_2 = df[['Sold_to_code','distribution_channel_description','Qty_BL_CM_Gross']]
    grouped10_2 = df10_2.groupby(['Sold_to_code','distribution_channel_description']).sum()
    grouped10_2 = grouped10_2.reset_index()
    grouped10_2_pivot = grouped10_2.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Qty_BL_CM_Gross').fillna(0)
    grouped10_2_pivot.columns = [''.join(col).strip() for col in grouped10_2_pivot.columns.values]
    grouped10_2_pivot = grouped10_2_pivot.reset_index()
    grouped10_2_pivot['prop_qty_B2B'] = grouped10_2_pivot.B2B/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot['prop_qty_STORE'] = grouped10_2_pivot.STORE/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot['prop_qty_STREET'] = grouped10_2_pivot.STREET/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot = grouped10_2_pivot.reset_index()
    #grouped10_2_pivot.head()
    grouped10_2_pivot_prop_vals = grouped10_2_pivot[['Sold_to_code','prop_qty_B2B','prop_qty_STORE','prop_qty_STREET']]
    feature10 = pd.merge(grouped10_1_pivot_prop_vals,grouped10_2_pivot_prop_vals)
    # Rename feature 10
    feature10.columns = ['Sold_to_code','amt_b2b_prop','amt_store_prop','amt_street_prop','qty_b2b_prop','qty_store_prop','qty_street_prop']
    
    # feature 11
    df11 = df[['Sold_to_code', 'distribution_channel_description', 'Amt_BL_CM_Gross']]
    split11 = df11.groupby(['Sold_to_code', 'distribution_channel_description'])
    avg_total_per_order = split11.mean().reset_index()
    ## Put NaN to 0 because it means no purchase...
    avg_total_per_order_pivot = avg_total_per_order.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Amt_BL_CM_Gross').reset_index()
    avg_total_per_order_pivot.columns = [''.join(col).strip() for col in avg_total_per_order_pivot.columns.values]
    avg_total_per_order_pivot = avg_total_per_order_pivot.fillna(0)
    avg_total_per_order_pivot.columns = ['Sold_to_code', 'amt_b2b_tot_avg', 'amt_store_tot_avg', 'amt_street_tot_avg']
    avg_total_per_order_pivot_transpose = avg_total_per_order_pivot.transpose()
    # avg_total_per_order_pivot_transpose['std_val_accross_channels']= 
    avg_total_per_order_pivot_transpose
    std_val_accross_channels = avg_total_per_order_pivot_transpose.loc[['amt_b2b_tot_avg', 'amt_store_tot_avg', 'amt_street_tot_avg'],:].std()
    avg_total_per_order_pivot['std_accross_channels_amt'] = std_val_accross_channels
    
    ## merge data
    data_2 = feature5.merge(feature_6,on='Sold_to_code').merge(feature7,on='Sold_to_code').merge(Feature_8,on='Sold_to_code').merge(Feature_9,on='Sold_to_code').merge(feature10,on='Sold_to_code').merge(avg_total_per_order_pivot,on='Sold_to_code')
    
    return data_2
```

```python
def CreateAllFeature(df):
    # We take a subset of our data frame, with only the needed variables
    # Each order :
    # - is associated with a unique Billing Document Number, 
    # - is placed by a Sold_to_code, 
    # - on a Shipment_Billing_Current_Analysis_Date
    # - has products which belongs to Material Groups & SubAxis
    # We sort the orders by increasing Shipment_Billing_Current_Analysis_Date and remove duplicates
    new_feature_1 = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 'Billing_Document_Number']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_1['previous_order_date'] = new_feature_1.groupby(['Sold_to_code'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # print(new_feature_1['previous_order_date'])
    # print(new_feature_1['Shipment_Billing_Current_Analysis_Date'])
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_1['lag'] = (new_feature_1['Shipment_Billing_Current_Analysis_Date'] - new_feature_1['previous_order_date']).dt.days
    # We groupb each 
    new_feature_1 = new_feature_1.groupby(["Sold_to_code"]).agg({'lag': ['min', 'max', 'mean', 'std']}).reset_index()
    new_feature_1.columns = [' '.join(col).strip() for col in new_feature_1.columns.values]
    # Rename feature 1
    new_feature_1.columns = ['Sold_to_code','min_time_diff','max_time_diff','avg_time_diff','std_time_diff']
    
    
    new_feature_2_MG = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 'Billing_Document_Number', 'material_group_Description']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date', 'material_group_Description']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_2_MG['previous_order_date'] = new_feature_2_MG.groupby(['Sold_to_code', 
                     'material_group_Description'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_2_MG['lag'] = (new_feature_2_MG['Shipment_Billing_Current_Analysis_Date'] - new_feature_2_MG['previous_order_date']).dt.days
    # We group each by customer and Material Group
    new_feature_2_MG = new_feature_2_MG.groupby(['Sold_to_code', 'material_group_Description']).agg({'lag': ['mean']}).reset_index()
    # We merge the two levels of column names
    new_feature_2_MG.columns = [' '.join(col).strip() for col in new_feature_2_MG.columns.values]
    # We pivot the table to get our MG in columns
    new_feature_2_MG = new_feature_2_MG.pivot_table(index=['Sold_to_code'],
                                                   columns=['material_group_Description'],
                                                   values='lag mean').reset_index()
    # We merge the two levels of column names
    new_feature_2_MG.columns = [''.join(col).strip() for col in new_feature_2_MG.columns.values]
    # Drop 2 columns
    new_feature_2_MG = new_feature_2_MG.drop(['Shopping Bags'], axis=1)

    # put columns names in lower function
    new_feature_2_MG.columns = ['Sold_to_code','accessory_promo_pos','allowance_funded_pac','bom_promo_no_fg','bom_sample_ppk_no_fg','bags',
                                'bandedpack_promo_gwp','box','displays_units_parts','fg_with_other_pack_i','financial_products','finished_goods',
                                'gwp','master_shipper_assor','other_pos','packaging_materials','partner_brand_fg','printed_materials','samples','t_shirts','testers','wip_basic_size']


    # Rename feature 2 Material Group without using first column
    new_names = [(i,i+'_time_reorder_avg') for i in new_feature_2_MG.iloc[:, 1:].columns.values]
    new_feature_2_MG.rename(columns = dict(new_names), inplace=True)
    
    
    new_feature_2_SA = df[['Shipment_Billing_Current_Analysis_Date', 'Sold_to_code', 
                               'Billing_Document_Number', 'Sub_Axe']]\
        .sort_values(['Sold_to_code','Shipment_Billing_Current_Analysis_Date', 
                     'Sub_Axe']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_2_SA['previous_order_date'] = new_feature_2_SA.groupby(['Sold_to_code', 
                     'Sub_Axe'])['Shipment_Billing_Current_Analysis_Date'].shift(1)
    # We calculate the lag expressed in days between each order and the previous one
    new_feature_2_SA['lag'] = (new_feature_2_SA['Shipment_Billing_Current_Analysis_Date'] - new_feature_2_SA['previous_order_date']).dt.days
    # We group each by customer and SubAxis
    new_feature_2_SA = new_feature_2_SA.groupby(['Sold_to_code', 'Sub_Axe']).agg({'lag': ['mean']}).reset_index()
    # We merge the two levels of column names
    new_feature_2_SA.columns = [' '.join(col).strip() for col in new_feature_2_SA.columns.values]
    # We pivot the table to get our MG in columns
    new_feature_2_SA = new_feature_2_SA.pivot_table(index=['Sold_to_code'],
                                                   columns=['Sub_Axe'],
                                                   values='lag mean').reset_index()
    # We merge the two levels of column names
    new_feature_2_SA.columns = [''.join(col).strip() for col in new_feature_2_SA.columns.values]

    # put columns names in lower function
    new_feature_2_SA.columns = ['Sold_to_code','haire_care','haire_color','others','styling']

    # Rename feature 2 SubAxis without using first column
    new_names_SubAxis = [(i,i+'_time_reorder_avg') for i in new_feature_2_SA.iloc[:, 1:].columns.values]
    new_feature_2_SA.rename(columns = dict(new_names_SubAxis), inplace=True)
    
    
    new_feature_3_value = df[['Sold_to_code', 'Sub_Axe', 'Amt_BL_CM_Gross']]\
        .sort_values(['Sold_to_code', 'Sub_Axe']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_3_value = new_feature_3_value.groupby(['Sold_to_code', 'Sub_Axe']).agg({'Amt_BL_CM_Gross': ['sum']}).reset_index()
    # We merge the two levels of column names
    new_feature_3_value.columns = [' '.join(col).strip() for col in new_feature_3_value.columns.values]

    # We compute a TEMPORARY total DataFrame
    tmp_total = df[['Sold_to_code', 'Amt_BL_CM_Gross']]\
        .sort_values(['Sold_to_code']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    tmp_total = tmp_total.groupby(['Sold_to_code']).agg({'Amt_BL_CM_Gross': ['sum']}).add_suffix('_total').reset_index()
    # We merge the two levels of column names
    tmp_total.columns = [' '.join(col).strip() for col in tmp_total.columns.values]

    # We merge both DataFrame
    new_feature_3_value = pd.merge(new_feature_3_value, tmp_total)
    # Divide the two columns
    new_feature_3_value['Amt_BL_CM_Gross_total_pct'] = new_feature_3_value['Amt_BL_CM_Gross sum'] / new_feature_3_value['Amt_BL_CM_Gross_total sum_total'] * 100
    # Drop the sum columns
    new_feature_3_value.drop(['Amt_BL_CM_Gross sum', 'Amt_BL_CM_Gross_total sum_total'], inplace=True, axis=1)
    # Remove the temporary DataFrame
    del tmp_total

    # We pivot the result data frame
    new_feature_3_value = new_feature_3_value.pivot_table(index=['Sold_to_code'],
                                                   columns=['Sub_Axe'],
                                                   values='Amt_BL_CM_Gross_total_pct').reset_index()

    # put columns names in lower function
    new_feature_3_value.columns = ['Sold_to_code','haire_care','haire_color','others','styling']

    # Rename feature 3 value per SUbAxis without using first column
    new_names_SubAxis_Value = [(i,'amt_per_SubAxis_'+i) for i in new_feature_3_value.iloc[:, 1:].columns.values]
    new_feature_3_value.rename(columns = dict(new_names_SubAxis_Value), inplace=True)
    
    
    new_feature_3_qty = df[['Sold_to_code', 'Sub_Axe', 'Qty_BL_CM_Gross']]\
        .sort_values(['Sold_to_code', 'Sub_Axe']).drop_duplicates()
    # We aggregate all orders by customers and SubAxis to get the sum of the values
    new_feature_3_qty = new_feature_3_qty.groupby(['Sold_to_code', 'Sub_Axe']).agg({'Qty_BL_CM_Gross': ['sum']}).reset_index()
    # We merge the two levels of column names
    new_feature_3_qty.columns = [' '.join(col).strip() for col in new_feature_3_qty.columns.values]

    # We compute a TEMPORARY total DataFrame
    tmp_total = df[['Sold_to_code', 'Qty_BL_CM_Gross']]\
        .sort_values(['Sold_to_code']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    tmp_total = tmp_total.groupby(['Sold_to_code']).agg({'Qty_BL_CM_Gross': ['sum']}).add_suffix('_total').reset_index()
    # We merge the two levels of column names
    tmp_total.columns = [' '.join(col).strip() for col in tmp_total.columns.values]

    # We merge both DataFrame
    new_feature_3_qty = pd.merge(new_feature_3_qty, tmp_total)
    # Divide the two columns
    new_feature_3_qty['Qty_BL_CM_Gross_total_pct'] = new_feature_3_qty['Qty_BL_CM_Gross sum'] / new_feature_3_qty['Qty_BL_CM_Gross_total sum_total'] * 100
    # Drop the sum columns
    new_feature_3_qty.drop(['Qty_BL_CM_Gross sum', 'Qty_BL_CM_Gross_total sum_total'], inplace=True, axis=1)
    # Remove the temporary DataFrame
    del tmp_total

    # We pivot the result data frame
    new_feature_3_qty = new_feature_3_qty.pivot_table(index=['Sold_to_code'],
                                                   columns=['Sub_Axe'],
                                                  values='Qty_BL_CM_Gross_total_pct').reset_index()

    # put columns names in lower function
    new_feature_3_qty.columns = ['Sold_to_code','haire_care','haire_color','others','styling']

    # Rename feature 3 quantity per SUbAxis without using first column
    new_names_SubAxis_qty = [(i,'qty_per_SubAxis_'+i) for i in new_feature_3_qty.iloc[:, 1:].columns.values]
    new_feature_3_qty.rename(columns = dict(new_names_SubAxis_qty), inplace=True)
    
    
    new_feature_4_value = df[['Sold_to_code', 'material_group_Description', 'Amt_BL_CM_Gross']]\
        .sort_values(['Sold_to_code', 'material_group_Description']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its value is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_4_value = new_feature_4_value.groupby(['Sold_to_code', 'material_group_Description']).agg({'Amt_BL_CM_Gross': ['sum']}).reset_index()
    # We merge the two levels of column names
    new_feature_4_value.columns = [' '.join(col).strip() for col in new_feature_4_value.columns.values]
    # We only select Material Groups where there are effective sales (amt > 0)
    new_feature_4_value = new_feature_4_value[new_feature_4_value['Amt_BL_CM_Gross sum'] > 0]
    # We select for each customer the amount of the least and most ordered item (in value)
    new_feature_4_value['count_max'] = new_feature_4_value.groupby(['Sold_to_code'])['Amt_BL_CM_Gross sum'].transform(max)
    new_feature_4_value['count_min'] = new_feature_4_value.groupby(['Sold_to_code'])['Amt_BL_CM_Gross sum'].transform(min)

    # Per customer, we get the first (in case of tie) MOST sold item
    # We only take the customer id and the material group
    new_feature_4_value_most = new_feature_4_value[['Sold_to_code', 'material_group_Description']]\
         [new_feature_4_value['Amt_BL_CM_Gross sum'] == new_feature_4_value['count_max']]\
        .groupby(['Sold_to_code']).first().rename(columns={"material_group_Description": "most_ordered_MG"})\
        .reset_index()

    # Per customer, we get the first (in case of tie) LEAST sold item
    # We only take the customer id and the material group
    new_feature_4_value_least = new_feature_4_value[['Sold_to_code', 'material_group_Description']]\
        [new_feature_4_value['Amt_BL_CM_Gross sum'] == new_feature_4_value['count_min']]\
        .groupby(['Sold_to_code']).first().rename(columns={"material_group_Description": "least_ordered_MG"})\
        .reset_index()

    # We merge both to get our results
    new_feature_4_value = pd.merge(new_feature_4_value_most, new_feature_4_value_least)
    # Deletion of the temporary DataFrame
    del new_feature_4_value_most, new_feature_4_value_least

    # put columns names in lower function
    new_feature_4_value.columns = ['Sold_to_code','most_ordered_mg','least_ordered_mg']

    # Rename feature 4 value per Material Group without using first column
    new_names_MG_value = [(i,'amt_'+i) for i in new_feature_4_value.iloc[:, 1:].columns.values]
    new_feature_4_value.rename(columns = dict(new_names_MG_value), inplace=True)
    
    
    new_feature_4_qty = df[['Sold_to_code', 'material_group_Description', 'Qty_BL_CM_Gross']]\
        .sort_values(['Sold_to_code', 'material_group_Description']).drop_duplicates()
    # We create a new column called previous_order_date
    # Its qty is the Shipment_Billing_Current_Analysis_Date during the previous order
    new_feature_4_qty = new_feature_4_qty.groupby(['Sold_to_code', 'material_group_Description']).agg({'Qty_BL_CM_Gross': ['sum']}).reset_index()
    # We merge the two levels of column names
    new_feature_4_qty.columns = [' '.join(col).strip() for col in new_feature_4_qty.columns.values]
    # We only select Material Groups where there are effective sales (Qty > 0)
    new_feature_4_qty = new_feature_4_qty[new_feature_4_qty['Qty_BL_CM_Gross sum'] > 0]
    # We select for each customer the amount of the least and most ordered item (in qty)
    new_feature_4_qty['count_max'] = new_feature_4_qty.groupby(['Sold_to_code'])['Qty_BL_CM_Gross sum'].transform(max)
    new_feature_4_qty['count_min'] = new_feature_4_qty.groupby(['Sold_to_code'])['Qty_BL_CM_Gross sum'].transform(min)

    # Per customer, we get the first (in case of tie) MOST sold item
    # We only take the customer id and the material group
    new_feature_4_qty_most = new_feature_4_qty[['Sold_to_code', 'material_group_Description']]\
         [new_feature_4_qty['Qty_BL_CM_Gross sum'] == new_feature_4_qty['count_max']]\
        .groupby(['Sold_to_code']).first().rename(columns={"material_group_Description": "most_ordered_MG"})\
        .reset_index()

    # Per customer, we get the first (in case of tie) LEAST sold item
    # We only take the customer id and the material group
    new_feature_4_qty_least = new_feature_4_qty[['Sold_to_code', 'material_group_Description']]\
        [new_feature_4_qty['Qty_BL_CM_Gross sum'] == new_feature_4_qty['count_min']]\
        .groupby(['Sold_to_code']).first().rename(columns={"material_group_Description": "least_ordered_MG"})\
        .reset_index()

    # We merge both to get our results
    new_feature_4_qty = pd.merge(new_feature_4_qty_most, new_feature_4_qty_least)
    # Deletion of the temporary DataFrame
    del new_feature_4_qty_most, new_feature_4_qty_least

    # put columns names in lower function
    new_feature_4_qty.columns = ['Sold_to_code','most_ordered_mg','least_ordered_mg']

    # Rename feature 4 quantity per Material Group without using first column
    new_names_qty = [(i,'qty_'+i) for i in new_feature_4_qty.iloc[:, 1:].columns.values]
    new_feature_4_qty.rename(columns = dict(new_names_qty), inplace=True)
    
    # feature 5
    df['qty*val*div'] = df.Qty_BL_CM_Gross*df.Amt_BL_CM_Gross*df.divertible
    df['qty*val'] = df.Qty_BL_CM_Gross*df.Amt_BL_CM_Gross
    df['qty*div'] = df.Qty_BL_CM_Gross*df.divertible
    df[['Qty_BL_CM_Gross','Amt_BL_CM_Gross','divertible','qty*val*div','qty*val', 'qty*div']].head()
    grouped5 = df.groupby('Sold_to_code')
    interdf51 = grouped5['qty*val*div','qty*val'].sum()
    interdf51['avg_div_in_value'] = interdf51['qty*val*div']/interdf51['qty*val']
    interdf51 = interdf51.reset_index()
    feature5_1 = interdf51[['Sold_to_code','avg_div_in_value']]
    interdf52 = grouped5['qty*div','Qty_BL_CM_Gross'].sum()
    interdf52['avg_div_in_qty'] = interdf52['qty*div']/interdf52['Qty_BL_CM_Gross']
    interdf52 = interdf52.reset_index()
    feature5_2 = interdf52[['Sold_to_code','avg_div_in_qty']]
    df5_3 = df[['Sold_to_code','Billing_Document_Number','Amt_BL_CM_Gross','divertible']]
    df5_3['div*Amt'] = df5_3.loc[:,'divertible']*df5_3.loc[:,'Amt_BL_CM_Gross']
    grouped5_3 = df5_3.groupby(['Sold_to_code','Billing_Document_Number']).sum()
    grouped5_3['max_prop_val_div'] = grouped5_3['div*Amt']/grouped5_3['Amt_BL_CM_Gross']
    inter5_3 = grouped5_3.reset_index()
    feature5_3 = inter5_3.groupby(['Sold_to_code']).max()
    feature5_3 = feature5_3['max_prop_val_div'].reset_index()
    df5_4 = df[['Sold_to_code','Billing_Document_Number','Qty_BL_CM_Gross','divertible']]
    df5_4['div*Qty'] = df5_4.loc[:,'divertible']*df5_4.loc[:,'Qty_BL_CM_Gross']
    grouped5_4 = df5_4.groupby(['Sold_to_code','Billing_Document_Number']).sum()
    grouped5_4['max_prop_qty_div'] = grouped5_4['div*Qty']/grouped5_4['Qty_BL_CM_Gross']
    inter5_4 = grouped5_4.reset_index()
    feature5_4 = inter5_4.groupby(['Sold_to_code']).max()
    feature5_4 = feature5_4['max_prop_qty_div'].reset_index()
    feature55 = pd.merge(feature5_1,feature5_2)
    feature56 = pd.merge(feature5_3,feature5_4)
    feature5 = pd.merge(feature55, feature56)
    # Rename feature 5
    feature5.columns = ['Sold_to_code','amt_divertible_in_avg','qty_divertible_in_avg','amt_divertible_prop_max','qty_divertible_prop_max']
    
    # feature 6
    df['diff_date_first_Now_td'] = (df.Shipment_Billing_Current_Analysis_Date - df.SC_First_Ship_Date)
    df['diff_date_first_Now'] = df['diff_date_first_Now_td'].dt.days
    #df['diff_date_first_Now'].head()
    grouped6 = df.groupby('Sold_to_code')
    feature_6 = grouped6['diff_date_first_Now'].agg(['min','max', 'mean']).reset_index()
    # Rename feature 6
    feature_6.columns = ['Sold_to_code','age_prod_avg','age_prod_max','age_prod_min']
    
    # feature 7
    df7_1 = df[['SC_First_Ship_Date','Shipment_Billing_Current_Analysis_Date','Sold_to_code','Amt_BL_CM_Gross']]
    grouped7_1 = df7_1.groupby(['Sold_to_code']).max()
    grouped7_1['age_most_prod_val'] = (grouped7_1.Shipment_Billing_Current_Analysis_Date - grouped7_1.SC_First_Ship_Date).dt.days
    grouped7_1 = grouped7_1.reset_index()
    df7_2 = df[['SC_First_Ship_Date','Shipment_Billing_Current_Analysis_Date','Sold_to_code','Qty_BL_CM_Gross']]
    grouped7_2 = df7_2.groupby(['Sold_to_code']).max()
    grouped7_2['age_most_prod_qty'] = (grouped7_2.Shipment_Billing_Current_Analysis_Date - grouped7_2.SC_First_Ship_Date).dt.days
    grouped7_2 = grouped7_2.reset_index()
    feature7 = pd.merge(grouped7_1, grouped7_2)
    feature7 = feature7[['Sold_to_code','age_most_prod_val','age_most_prod_qty']]
    # Rename feature 7
    feature7.columns = [['Sold_to_code','amt_age_most_prod_avg','qty_age_most_prod_avg']]    
    
    # feature 8
    ## Nb distinct 'ship' adresses for not STORE
    split8 = df[df.distribution_channel_description != 'STORE'].groupby('Sold_to_code')
    disct_ship_not_store = split8.street_address.nunique()
    feature8 = pd.DataFrame(disct_ship_not_store).reset_index()
    all_stc = set(new_feature_1.Sold_to_code) # all data with column name Sold_to_code
    some_stc_ft_8 = set(disct_ship_not_store.index) # data with 1
    missing_ft_8 = all_stc - some_stc_ft_8 # missing data
    # je remplace des valeurs manquantes par des zeros Feature 8
    S_ft_8 = list(missing_ft_8)
    L_ft_8 = []
    for i in range(len(S_ft_8)):
        if S_ft_8[i] > 1:
            L_ft_8.append(0)
    #L_ft_8
    # je rajoute des zero remplacs prcdemment dans some_stc_ft_8
    List_feature_8_column_address = list(feature8['street_address'])
    List_feature_8_column_address.extend(L_ft_8)
    # je rajoute les Sold_to_code qui manquent dans features 8 colonne Sold_to_code
    List_feature_8_column_Sold = list(feature8['Sold_to_code'])
    List_feature_8_column_Sold.extend(S_ft_8)
    # DataFrame
    Feature_8 = pd.DataFrame({'Sold_to_code': List_feature_8_column_Sold, 
                          'street_address_channel_SC': List_feature_8_column_address})
    
    
    # feature 9
    ## Nb distinct 'ship' adresses for STORE
    split9 = df[df.distribution_channel_description == 'STORE'].groupby('Sold_to_code')
    disct_ship_store = split9.street_address.nunique()
    feature9 = pd.DataFrame(disct_ship_store).reset_index()
    
    all_stc = set(new_feature_1.Sold_to_code) # all data with column name Sold_to_code
    some_stc_ft_9 = set(disct_ship_store.index) # data with 1
    missing_ft_9 = all_stc - some_stc_ft_9 # missing data
    # je remplace des valeurs manquantes par des zeros Feature 9
    S_ft_9 = list(missing_ft_9)
    L_ft_9 = []
    for i in range(len(S_ft_9)):
        if S_ft_9[i] > 1:
            L_ft_9.append(0)
    #L_ft_9
    # je rajoute des zero remplacs prcdemment dans some_stc_ft_9
    List_feature_9_column_address = list(feature9['street_address'])
    List_feature_9_column_address.extend(L_ft_9)
    #List_feature_9_column_address
    # je rajoute les Sold_to_code qui manquent dans features 9 colonne Sold_to_code
    List_feature_9_column_Sold = list(feature9['Sold_to_code'])
    List_feature_9_column_Sold.extend(S_ft_9)
    #List_feature_9_column_Sold
    # DataFrame
    Feature_9 = pd.DataFrame({'Sold_to_code': List_feature_9_column_Sold, 
                          'street_address_channel_not_SC': List_feature_9_column_address})
    
    # feature 10
    df10_1 = df[['Sold_to_code','distribution_channel_description','Amt_BL_CM_Gross']]
    grouped10_1 = df10_1.groupby(['Sold_to_code','distribution_channel_description']).sum()
    grouped10_1 = grouped10_1.reset_index()
    grouped10_1_pivot = grouped10_1.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Amt_BL_CM_Gross').fillna(0)
    grouped10_1_pivot.columns = [''.join(col).strip() for col in grouped10_1_pivot.columns.values]
    grouped10_1_pivot = grouped10_1_pivot.reset_index()
    grouped10_1_pivot['prop_val_B2B'] = grouped10_1_pivot.B2B/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot['prop_val_STORE'] = grouped10_1_pivot.STORE/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot['prop_val_STREET'] = grouped10_1_pivot.STREET/(grouped10_1_pivot.B2B + grouped10_1_pivot.STORE + grouped10_1_pivot.STREET)
    grouped10_1_pivot = grouped10_1_pivot.reset_index()
    grouped10_1_pivot_prop_vals = grouped10_1_pivot[['Sold_to_code','prop_val_B2B','prop_val_STORE','prop_val_STREET']]
    df10_2 = df[['Sold_to_code','distribution_channel_description','Qty_BL_CM_Gross']]
    grouped10_2 = df10_2.groupby(['Sold_to_code','distribution_channel_description']).sum()
    grouped10_2 = grouped10_2.reset_index()
    grouped10_2_pivot = grouped10_2.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Qty_BL_CM_Gross').fillna(0)
    grouped10_2_pivot.columns = [''.join(col).strip() for col in grouped10_2_pivot.columns.values]
    grouped10_2_pivot = grouped10_2_pivot.reset_index()
    grouped10_2_pivot['prop_qty_B2B'] = grouped10_2_pivot.B2B/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot['prop_qty_STORE'] = grouped10_2_pivot.STORE/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot['prop_qty_STREET'] = grouped10_2_pivot.STREET/(grouped10_2_pivot.B2B + grouped10_2_pivot.STORE + grouped10_2_pivot.STREET)
    grouped10_2_pivot = grouped10_2_pivot.reset_index()
    #grouped10_2_pivot.head()
    grouped10_2_pivot_prop_vals = grouped10_2_pivot[['Sold_to_code','prop_qty_B2B','prop_qty_STORE','prop_qty_STREET']]
    feature10 = pd.merge(grouped10_1_pivot_prop_vals,grouped10_2_pivot_prop_vals)
    # Rename feature 10
    feature10.columns = ['Sold_to_code','amt_b2b_prop','amt_store_prop','amt_street_prop','qty_b2b_prop','qty_store_prop','qty_street_prop']
    
    # feature 11
    df11 = df[['Sold_to_code', 'distribution_channel_description', 'Amt_BL_CM_Gross']]
    split11 = df11.groupby(['Sold_to_code', 'distribution_channel_description'])
    avg_total_per_order = split11.mean().reset_index()
    ## Put NaN to 0 because it means no purchase...
    avg_total_per_order_pivot = avg_total_per_order.pivot(index='Sold_to_code', columns='distribution_channel_description', values='Amt_BL_CM_Gross').reset_index()
    avg_total_per_order_pivot.columns = [''.join(col).strip() for col in avg_total_per_order_pivot.columns.values]
    avg_total_per_order_pivot = avg_total_per_order_pivot.fillna(0)
    avg_total_per_order_pivot.columns = ['Sold_to_code', 'amt_b2b_tot_avg', 'amt_store_tot_avg', 'amt_street_tot_avg']
    avg_total_per_order_pivot_transpose = avg_total_per_order_pivot.transpose()
    # avg_total_per_order_pivot_transpose['std_val_accross_channels']= 
    avg_total_per_order_pivot_transpose
    std_val_accross_channels = avg_total_per_order_pivot_transpose.loc[['amt_b2b_tot_avg', 'amt_store_tot_avg', 'amt_street_tot_avg'],:].std()
    avg_total_per_order_pivot['std_accross_channels_amt'] = std_val_accross_channels

    # merge all features
    data = new_feature_1.merge(new_feature_2_MG,on='Sold_to_code').merge(new_feature_2_SA,on='Sold_to_code').merge(new_feature_3_qty,on='Sold_to_code').merge(new_feature_3_value,on='Sold_to_code').merge(new_feature_4_qty,on='Sold_to_code').merge(new_feature_4_value,on='Sold_to_code').merge(feature5,on='Sold_to_code').merge(feature_6,on='Sold_to_code').merge(feature7,on='Sold_to_code').merge(Feature_8,on='Sold_to_code').merge(Feature_9,on='Sold_to_code').merge(feature10,on='Sold_to_code').merge(avg_total_per_order_pivot,on='Sold_to_code')
    
    return data
```
